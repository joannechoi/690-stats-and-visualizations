**Duck Typing Example**

````
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError: # not iterable
        return False

isiterable('a string')
````

````
isiterable([1, 2, 3])
````

````
isiterable(5)
````

````
if not isinstance(x, list) and isiterable(x):
    x = list(x)
````

**if, elif, and else Examples**

````
x = 9

if x < 0:    print ('It\'\s negative')
elif x == 0:
    print ('Equal to zero')
elif 0 < x < 5:
    print ('Positive but smaller than 5')
else:
    print ('Positive and larger than or equal to 5')
````

````
a = 5; b = 7

c = 8; d = 4

if a < b or c > d: # c > d does not get evaluated because a < b is True
    print ('Made it')
````

**loops Examples**

````
x = 256
total = 0
while x > 0:
    if total > 500:
        break 
    total += x
    x = x // 2

print(x)
````
