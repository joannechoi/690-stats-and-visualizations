#### Step 1: Create a 10x10 matrix with random integers between 0 and 9

```
import random 

matrix = []

for row in range(10):
    values = []
    for col in range(10):
        values.append(random.randrange(10))
    matrix.append(values)
    print(*values)
```

#### Step 2: Replace all odd numbers with "@"

```
for i in matrix:
        odd = ["@" if x % 2 != 0 else x for x in i]
        print(*odd)
```

#### Step 3: Calculate the sum of each row.
Note: Display the values in each row and sum, separated by "*"

```
total_row = list(map(sum, matrix))

for m, r in zip(matrix, total_row):
    print(*m, "*", r)
```

#### Step 4: Create a border around the matrix

```
print("*" * 23)
for m, in zip(matrix):
    print("*", *m, "*")
print("*" * 23)
```

#### Step 5: Display the sum of each row and column. 
Note: Display the sum of row to the right of the matrix and display the sum of column at the bottom of the matrix. 

```
total_column = list(map(sum, zip(*matrix)))

for m, r, in zip(matrix, total_row):
    print(*m, "*", r, sep='  ')
print("*  "* 10)
print(*total_column)
```
