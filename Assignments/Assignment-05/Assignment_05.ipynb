2. Create a smaller data frame with following 9 variables: 
- **UNITID** - Unit ID for institution
- **INSTNM** - Institution name
- **CITY** - City
- **STABBR** - State postcode
- **ZIP** - Zip Code
- **ADM_RATE** - Admission rate
- **UGDS** - Enrollment of undergraduate certificate/degree-seeking students
- **TUITIONFEE_IN** - In-state tuition and fees
- **MN_EARN_WNE_P10** - Mean earnings of students working and not enrolled 10 years after entry

```
colleges = pd.DataFrame(data, columns = ['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10'])
```

- Display the first 10 colleges.

```
colleges.head(10)
```

 - Display the last 10 colleges.

```
colleges.tail(10)
```

- Display 10 random colleges. 

```
colleges.sample(10)
```

```
md = colleges[colleges['STABBR'].str.contains('MD')]
md
```

- Display information for UMBC.

```
umbc = md[md['INSTNM'] == 'University of Maryland-Baltimore County']

umbc
```

- Display information for Univ. of Maryland College Park.

```
umd = md[md['INSTNM'] == 'University of Maryland-College Park']

umd
```

- Display information for Loyola College Maryland.

```
loyola = md[md['INSTNM'] == 'Loyola University Maryland']

loyola
```

- Display information for Johns Hopskins. 

```
jh = md[md.INSTNM.str.contains('Johns Hopkins')]

jh
```

- Create a dataframe of the above colleges. 

```
sub_colleges = umbc.append([umd, loyola, jh])

sub_colleges
```

```
sub_colleges['ADM_RATE'].mean()
```

3.
Filter the smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia. 

```
states = ['MD', 'VA', 'DC', 'PA', 'DE']

states_colleges = colleges[colleges['STABBR'].isin(states)]
states_colleges
```

- Save the filtered dataset to a new file named "Five_States_Colleges.csv" for future analysis.

```
states_colleges.to_csv('Five_States_Colleges.csv', index = False)

new_df = pd.read_csv('Five_States_Colleges.csv')

new_df.head()
```
