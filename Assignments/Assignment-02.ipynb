# Assignment #2

## Requirements:
 - Prompts a user to enter 10 integers.
 - If the user enters anything other than integers, remind her that only integers are allowed and let her retry.
   - Note: You need to detect if use makes a mistake and let the user retry without exiting the program. The previously entered valid inputs should be kept and not lost.
 - Don't allow the user to enter more than 10 or less than 10 integers.
 - Display the 10 integers back to the user at the end.
 - Calculate the following statistics from the 10 integers entered and display the results back to the user:
   - Minimum
   - Maximum
   - Range
   - Mean
   - Variance
   - Standard Deviation

### Step #1: Prompt user to enter 10 integers.
 - Do not allow user to enter non-integer values
 - Retain allow `True` values into a list
 - Display the final set of integers back to the user.

````
user_inputs = []
count = 10

for i in range (0,10):
    while True:
        try:
            user_input = int(input("Enter a number then press the 'Enter' button."))
            user_inputs.append(user_input)
            count -= 1
            break
        except: 
            print("The value you have entered is not a number. Please enter a number.")
            
    print("The value you have entered is:", user_input)
    
    print("Please repeat this step", count, "more times.")

print("You have entered the following set of numbers:", user_inputs)
````

### Step #2: Calculate various statistics with the user data.
- Find the minimum value
- Find the maximum value
- Find the range
- Find the mean
- Find the variance
- Find the standard deviation

#### Calculating the Minimum
1. Find the lowest value in the set. 

````
min_value = user_inputs[0]

for i in user_inputs:
    if i < min_value:
        min_value = i

print("The minimum value in the set is:", min_value)
````

#### Calculating the Maximum
1. Find the highest value in the set. 

````
max_value = user_inputs[0]

for i in user_inputs:
    if i > max_value:
        max_value = i

print("The maximum value in the set is:", max_value)
````

#### Calculating the Range
1. Find the minimum value.
2. Find the maximum value.
3. Subtract minimum value from maximum value.

```
maximum value - minimum value = range

````
range_value = max_value - min_value

print("The range of the set is:", range_value)
````

#### Calculating the Mean
1. Add up all of the values in the set.
2. Divide by the count of values in the set.

```
sum of all value / count of values = mean
```

````
input_sum = sum(user_inputs)
input_total = len(user_inputs)
input_avg = input_sum // input_total

print("The mean of the set is:", input_avg)
````

#### Calculating the Variance
1. Calculate the mean.
2. Find the difference of each value in the set from the mean. 
```
(i1 - mean) + (i2 - mean) + (i3 - mean) + (in - mean) = difference from mean
```
3. Square each difference.
```
[(i1 - mean)**2] + [(i2 - mean)**2] + [(i3 - mean)**2] + [(in - mean)**2] = squared difference from mean
```
4. Take the squared difference and find the mean.
```
squared difference from mean / count of the set = variance
```

````
input_avg_diff = 0

for i in range(input_total):
    input_avg_diff= input_avg_diff + (user_inputs[i] - input_avg)**2

input_variance = input_avg_diff / input_total

print("The variance of the set is:", input_variance)
````

#### Calculate the Standard Deviation
1. Find the square root of the variance

````
input_sd = input_variance ** 0.5

print("The standard deviation of the set is:", input_sd)
````
